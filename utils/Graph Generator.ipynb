```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
csv_path = ''
```

```python
def plot_line_graph(csv_path, y='Value', lower_range_name='__MIN', upper_range_name='__MAX', group=''):
    filtered_terms = ['step', lower_range_name, upper_range_name]
    filtered_terms_l = [x.lower() for x in filtered_terms]

    df = pd.read_csv(csv_path)

    # Get the distinct name of data
    filtered_columns = [col for col in df.columns if not any(col.lower().endswith(term) for term in filtered_terms_l)]

    df_melted = pd.melt(df, id_vars=['Step'], value_vars=filtered_columns, var_name=group, value_name=y)

    # Style the plot
    sns.set_style()
    sns.lineplot(data=df_melted, x='Step', y=y, hue='')
    for group in filtered_columns:
        plt.fill_between(df['Step'], df[f'{group}{lower_range_name}'], df[f'{group}{upper_range_name}'], alpha=0.2)
```

```python
plot_line_graph(csv_path, y='SWAG')
```
